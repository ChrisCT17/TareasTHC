# **Leyendas y guardar graficas**

```python
import matplotlib.pyplot as plt #importamos las bibliotecas
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])
plt.xlim(0, 4) #Nos da la grafica hasta un cierto limite, es decir, acotamos la grafica
plt.ylim(0, 200)
plt.xticks(fontsize=10) #xticks y fontsize cambia el tama√±o de los numeros sobre los ejes en el grafico
plt.yticks(fontsize=10)
plt.legend(title="Funcion lineal", loc="upper left", fontsize=15) #Legend genera una leyenda en el grafico y loc la coloca en una posicion especifica

plt.savefig("funcion lineal.png", dpi=300, bbox_inches="tight") #Guarda la grafica para descargarla o invocarla mas sencillamente

plt.plot(xpoints, ypoints)

plt.show()
```

# **Ejercicio**

```python
#Subgrafica 1
x = np.linspace(-10,10,100)
y=np.sin(x*2)
plt.subplot(2,2,1)
plt.title("Funcion seno")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid()
plt.plot(x,y, color="r")

#Subgrafica 2
x=np.linspace(-20,20,100)
y=np.tanh(x)
plt.subplot(2,2,2)
plt.title("Funcion tanh")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid()
plt.plot(x,y,color="c")

#Subgrafica 3
x=np.linspace(-10,10,100)
y=np.sin(x)/x
y[np.where(x==0)]=1
plt.subplot(2,2,3)
plt.title("Funcion de seno sobre x")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid()
plt.plot(x,y,color="y")

#subgrafica 4
x=np.linspace(-0,50,100)
y=np.exp(-x/10)
plt.subplot(2,2,4)
plt.title("Funcion exponencial")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid()
plt.plot(x,y,color="b")

plt.tight_layout()
plt.show()
```
