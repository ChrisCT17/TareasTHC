```python
x = np.linspace(0,20,100) #Definimos x con linspace, va desde el 0 al 20 y genera 100 valores entre estos
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1) #graficamos ambas graficas, puede ser en una misma linea
plt.plot(x, y2)
plt.show()
```

```python
x = np.linspace(0,20,100) #Definimos x con linspace, va desde el 0 al 20 y genera 100 valores entre estos
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1,"o") #Nos da la grafica con puntos
plt.plot(x, y2,"*") #Nos da la grafica con formas de estrella
plt.show()
```

```python
x = np.linspace(0,20,100) #Definimos x con linspace, va desde el 0 al 20 y genera 100 valores entre estos
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1,"r") #Nos da la grafica de color rojo, utilizando la inicial del color, en este caso red
plt.plot(x, y2,"g") #Nos da la grafica de color verde (green)
plt.show()
```

```python
x = np.linspace(0,20,100) #Definimos x con linspace, va desde el 0 al 20 y genera 100 valores entre estos
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, marker="o") #Nos da la grafica con puntos unidos por lineas, lo mismo pasa con otros caracteres como *
plt.plot(x, y2, marker="*")
plt.show()
```

```python
x = np.linspace(0,20,100) #Definimos x con linspace, va desde el 0 al 20 y genera 100 valores entre estos
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, marker="D")
plt.plot(x, y2, marker="4")
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10]) #Definimos un arreglo de numeros, como no definimos x, estos valores se asignan automaticamente desde el 0

plt.plot(ypoints,"o:r") #Modifica la linea y el color, o representa los puntos y r el color que en este caso es rojo. Los dos puntos cambian la linea
plt.show()
```

```python
x=np.linspace(0,20,10)
y1=x*2
plt.plot(x,y1,"r-.") #Esto tambien modifca la linea, r cambia el color a rojo, -. definen al tipo de linea
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = "o", ms=20) #ms es el tamao del punto, es abreviatura de markersize
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = "o", ms=20, mec= "r") #mec le da un contorno de diferente color a los puntos
plt.show()
```

```python
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y,linestyle="-",color="c")
plt.xlabel("Eje x") #Etiqueta en el eje x
plt.ylabel("Eje y") #Etiqueta en el eje y
plt.title("Funcion cuadratica") #Titulo de la grafica
plt.show()
```

**Subplots: Genera dos graficas en un mismo espacio**

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) #el primer valor es la entrada, el segundo la columna y el tercero la posicion
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
font1 = {'family':'serif','color':'blue','size':10} #Definimos el primer formato
font2 = {'family':'serif','color':'purple','size':10} #Definimos el segundo formato
plt.xlabel("Eje x",fontdict=font1) #Etiqueta en el eje x, modificamos su formato
plt.ylabel("Eje y",fontdict=font1) #Etiqueta en el eje y
plt.title("Funcion cuadratica y seno",fontdict=font2) #Titulo de la grafica, modificamos su formato
#plot 1:
x = np.linspace(10,50,100)
y = x**2
plt.subplot(1, 2, 1)

plt.plot(x,y)

#plot 2:
y=np.sin(x)
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.show()
```
